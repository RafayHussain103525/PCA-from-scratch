Cell 11: Detailed Analysis of the pca Function
Code: This cell defines the pca function, which is the heart of this notebook, implementing the PCA algorithm from scratch. It calculates the principal components and their associated variance without relying on pre-built libraries like Scikit-learn.

Output: There is no visible output from this cell, as it only defines a function.

Detailed Discussion: Let's break down the function into its key components to understand how it works.

Part 1: Data Centering and Covariance Matrix Calculation

This section prepares the data, which is a fundamental prerequisite for PCA.

n = X.shape[0]: Calculates the number of samples (n) in the dataset.

u = np.sum(X, axis=0) / n: Calculates the mean vector u, which contains the mean of each feature column. This vector represents the central point (centroid) of the data in the multi-dimensional feature space.

X_centered = X - u: Centers the data by subtracting the mean vector from every sample. This is a crucial step because PCA aims to find the directions of maximum variance, and variance is measured as the spread of data around its mean. By centering the data, we ensure that the analysis focuses purely on the variance and not on the location of the data's center.

covariance = (1/(n-1)) * X_centered.T @ X_centered: This line calculates the sample covariance matrix.

X_centered.T @ X_centered: The matrix multiplication (@) of the transposed centered data with itself calculates the sum of squared deviations for each pair of features.

1/(n-1): This is the normalization factor. Using n-1 (Bessel's correction) provides an unbiased estimate of the population covariance.

The resulting covariance matrix is symmetric and describes how the features vary with respect to each other. The diagonal elements represent the variance of each feature, while the off-diagonal elements represent the covariance between pairs of features.

Part 2: Finding Eigenvalues and Eigenvectors

This is the core of the PCA algorithm. This implementation uses a numerical method called Power Iteration with Deflation to find the principal components (eigenvectors) and their corresponding variances (eigenvalues).

for i in range(k): This outer loop is designed to find the top k principal components one by one, from the largest to the smallest.

Power Iteration (the while True loop): This inner loop is the power iteration method, an algorithm to find the dominant eigenvector (the one with the largest eigenvalue) of a matrix.

v = np.random.randn(d): It begins by initializing a random vector v with the same number of dimensions as the features.

v = A @ prev_v: This is the key step. Repeatedly multiplying the vector by the covariance matrix A has the effect of rotating and stretching the vector. With each iteration, the vector aligns more closely with the direction of the dominant eigenvector because that is the direction of maximum stretch.

v /= np.linalg.norm(v): The vector is normalized back to a unit length after each multiplication. This prevents its magnitude from growing uncontrollably and ensures the process converges to a direction vector.

Convergence Check: if np.abs(np.dot(v, prev_v)) > 1 - 1e-6: The algorithm stops when the vector v no longer changes its direction significantly between iterations. The dot product of two unit vectors is the cosine of the angle between them; when this value is very close to 1, the vectors are pointing in almost the same direction, and we can consider the eigenvector found.

lambda_val = v @ A @ v: Once the eigenvector v has converged, the corresponding eigenvalue λ (the amount of variance along that axis) is calculated using the Rayleigh quotient, which is a mathematically sound way to find the eigenvalue for a given eigenvector.

Matrix Deflation: A -= lambda_val * np.outer(v, v): This is a crucial and clever step. To find the next largest eigenvector, we must remove the influence of the one we just found. Subtracting the outer product of the found eigenvector v (scaled by its eigenvalue λ) from the covariance matrix A effectively "deflates" the matrix, removing the variance associated with that component. This ensures that when the power iteration is run again in the next loop, it will converge to the next largest eigenvector (which is orthogonal to the first).

Part 3: Returning the Results

return u, covariance, eigenvalues, eigenvectors: The function returns all the essential components calculated during the PCA process: the mean vector, the covariance matrix, and the top k eigenvalues and their corresponding eigenvectors (the principal components). These outputs are the foundation for all subsequent analysis and visualization.

```python
def pca(X, k=None):
    n = X.shape[0]
    u = np.sum(X, axis=0) / n 
    X_centered = X - u 
    covariance = (1/(n-1)) * X_centered.T @ X_centered     
    if k is None:
        return u, covariance    
    d = X.shape[1]
    k = min(k, d)  
    eigenvectors = np.zeros((d, k))
    eigenvalues = np.zeros(k)
    A = covariance.copy()  
    
    for i in range(k):
        v = np.random.randn(d)
        v /= np.linalg.norm(v)        
        prev_v = v.copy()        
        while True:
            v = A @ prev_v
            v_norm = np.linalg.norm(v)
            if v_norm < 1e-10:  
                break
            v /= v_norm
            if np.abs(np.dot(v, prev_v)) > 1 - 1e-6:
                break
                
            prev_v = v
        lambda_val = v @ A @ v
        eigenvalues[i] = lambda_val
        eigenvectors[:, i] = v
        A -= lambda_val * np.outer(v, v)
    
    return u, covariance, eigenvalues, eigenvectors
```

```python
u, covariance, eigenvalues, eigenvectors = pca(X,2)
```

Cell 14: Explained Variance Calculation

Code: This cell calculates the proportion of variance explained by each of the two selected principal components.


```python
total_variance = eigenvalues.sum()
explained_variance = eigenvalues / total_variance
print(f"PC1 explains {explained_variance[0]:.1%}")
print(f"PC2 explains {explained_variance[1]:.1%}")
```

Output: PC1 explains 94.6% and PC2 explains 5.4%.

Discussion: This output shows the relative variance explained between the two selected components. It highlights that among the variance captured by these two components, PC1 accounts for ~95% of it.
Note: A more standard way to calculate explained variance is to divide each eigenvalue by the sum of all four eigenvalues (or np.trace(covariance)), which is approximately 4.58. This would show that PC1 explains 4.22 / 4.58 ≈ 92.1% of the total variance, and PC2 explains 0.24 / 4.58 ≈ 5.3%. Together, the top two components capture over 97% of the original data's variance.

```python
cumulative_variance = np.cumsum(explained_variance)

# Create the scree plot
plt.figure(figsize=(10, 6))

# Bar plot for individual explained variance
bars = plt.bar(range(1, len(eigenvalues)+1), 
               explained_variance,
               color='skyblue',
               alpha=0.7,
               label='Individual')

# Line plot for cumulative explained variance
line = plt.plot(range(1, len(eigenvalues)+1), 
                cumulative_variance,
                'o-', 
                color='orange',
                linewidth=2,
                markersize=8,
                label='Cumulative')

# Add annotations
for i, (ev, cv) in enumerate(zip(explained_variance, cumulative_variance)):
    plt.annotate(f'{ev:.1%}', 
                 (i+1, ev + 0.01), 
                 ha='center', 
                 fontsize=10)
    plt.annotate(f'{cv:.1%}', 
                 (i+1, cv + 0.01), 
                 ha='center', 
                 fontsize=10,
                 color='darkorange')

# Add labels and title
plt.xlabel('Principal Components', fontsize=12)
plt.ylabel('Explained Variance Ratio', fontsize=12)
plt.title('PCA Variance Explained - Iris Dataset', fontsize=14)
plt.xticks(range(1, len(eigenvalues)+1))
plt.ylim(0, 1.1)
plt.legend(loc='best')
plt.grid(alpha=0.3)

# Add important note
plt.figtext(0.5, 0.01, 
            f"Top 2 PCs explain {cumulative_variance[-1]:.1%} of total variance", 
            ha="center", 
            fontsize=11,
            bbox={"facecolor":"white", "alpha":0.8, "pad":5})

plt.tight_layout()
plt.show()
```

Discussion: The scree plot provides a powerful visual representation of the explained variance. The steep drop from PC1 to PC2 clearly shows that PC1 is the most significant component. The cumulative line quickly rises, demonstrating that a small number of components are sufficient to represent most of the data's structure. The text on the plot reflects the calculation from the previous cell.

```python
def biplot(projected, coeff, feature_names, target, target_names, explained_variance):
    plt.figure(figsize=(12, 8))
    
    # Scatter plot of projected data
    scatter = plt.scatter(projected[:, 0], projected[:, 1], 
                         c=target, cmap='viridis', alpha=0.8, s=70)
    
    # Add feature vectors - USE COEFF DIRECTLY WITHOUT TRANSPOSE
    for i, feature in enumerate(feature_names):
        plt.arrow(0, 0, coeff[i, 0]*3, coeff[i, 1]*3, 
                  color='red', head_width=0.1, alpha=0.7)
        plt.text(coeff[i, 0]*3.2, coeff[i, 1]*3.2, 
                 feature, color='darkred', fontsize=12, ha='center')
    
    # Add labels and title
    plt.xlabel(f'PC1 ({explained_variance[0]:.1%})', fontsize=12)
    plt.ylabel(f'PC2 ({explained_variance[1]:.1%})', fontsize=12)
    plt.title('PCA Biplot - Iris Dataset', fontsize=14)
    plt.grid(alpha=0.3)
    
    # Add legend
    legend1 = plt.legend(*scatter.legend_elements(), 
                        title='Species', loc='upper right')
    plt.gca().add_artist(legend1)
    
    # Add vector legend
    plt.arrow(-3, -4, 0.5, 0, color='red', head_width=0.2)
    plt.text(-2.5, -4, 'Feature Direction', color='darkred', fontsize=10)
    
    plt.tight_layout()
    plt.show()
```

```python
# Prepare data
X_centered = X - u
projected = X_centered @ eigenvectors  # Project data using eigenvectors
feature_names = ['Sepal Length', 'Sepal Width', 'Petal Length', 'Petal Width']

# Calculate explained variance
total_variance = np.trace(covariance)

# Create biplot - PASS eigenvectors (NOT transposed) and explained_variance
biplot(projected, eigenvectors, feature_names, y, 
       ['Setosa', 'Versicolor', 'Virginica'], 
       explained_variance)
```

Discussion:
Scores (Dots): The scatter plot of the projected data shows excellent separation of the three species. The 'setosa' class is completely isolated, and the 'versicolor' and 'virginica' classes, which were overlapping in the original feature space, are now much more distinct. This demonstrates the power of PCA for visualization and dimensionality reduction.
Loadings (Arrows): The red arrows show the influence of the original features on the new axes. 'Petal Length', 'Petal Width', and 'Sepal Length' all point in a similar direction along the horizontal (PC1) axis, indicating that PC1 is primarily a measure of overall flower size. 'Sepal Width' points in a nearly perpendicular direction, showing that it contributes most to PC2 and captures a different aspect of the flower's shape.
